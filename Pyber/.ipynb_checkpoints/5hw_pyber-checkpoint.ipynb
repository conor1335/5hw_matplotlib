```python
#%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
cd = pd.read_csv(city_data_to_load)
rd = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset

# Display the data table for preview
cr = pd.merge(rd, cd, on="city", how="left")
cr.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
crc = cr.groupby(['city'])
x = cr['city'].value_counts()
per_fare = crc.sum()['fare']
y = crc.mean()['fare']
z = crc.mean()['driver_count']
w = crc['type']
b = w.unique()
a = pd.DataFrame({
    'x': x,
    'y': y,
    'z': z,
    'b': b,
    'per_fare': per_fare
})
```

```python
urban_only = a.loc[a["b"] == "Urban", :]
```

```python
suburban_only = a.loc[a["b"] == "Suburban", :]

```

```python
rural_only = a.loc[a["b"] == "Rural", :]
```

```python
urbs_dots = plt.scatter(urban_only['x'], urban_only['y'],  marker="o", label='Urban',
                        facecolors="lightcoral", edgecolors="black", s = 10*urban_only['z'], alpha=0.8, linewidth=1)
subs_dots = plt.scatter(suburban_only['x'], suburban_only['y'],  marker="o", label='Suburban',
                        facecolors="lightskyblue", edgecolors="black", s = 10*suburban_only['z'], alpha=0.8,linewidth=1)
rurs_dots = plt.scatter(rural_only['x'], rural_only['y'],  marker="o", label='Rural',
                        facecolors="yellow", edgecolors="black", s = 10*rural_only['z'], alpha=0.58,linewidth=1)

lgnd = plt.legend(handles=[urbs_dots, subs_dots, rurs_dots], loc="best", 
scatterpoints=1, fontsize=10, title="City Types")
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

plt.savefig("Image/pyber_ride_share.png")

plt.grid()


# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend
plt.text(43, 35,"Note:\nCircle size correlates with\ndriver count per city")

# Incorporate a text label regarding circle size

# Save Figure
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urb_num = urban_only['per_fare'].sum()
sub_num = suburban_only['per_fare'].sum()
rur_num = rural_only['per_fare'].sum()

sizes = [urb_num, sub_num, rur_num]

labels = ["Urban", "Suburban", "Rural"]
colors = ["lightcoral", "lightskyblue", 'yellow']
explode = (0, 0.1, 0.1)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True)
plt.title("% of Total Fares by City Type")

plt.axis("equal")


plt.savefig("Image/fares_by_city_type.png")
# Build Pie Chart

# Save Figure
```
